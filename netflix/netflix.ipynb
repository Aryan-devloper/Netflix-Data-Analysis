catagorizing Vote_average



```python
def catigorize_col(df, col, labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]
    
    df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df

labels = ['not_popular', 'below_avg', 'average', 'popular']
df = catigorize_col(df, 'Vote_Average', labels)
print(df['Vote_Average'].unique())
```

```python
df.dropna(inplace=True)
print(df.isna().sum())
```

1.What is the most Frequent genre in the dataset ?

```python
df = df.dropna(subset=['Genre'])
all_genres = df['Genre'].str.split(', ').explode()
genre_counts = all_genres.value_counts()
plt.barh(genre_counts.index[:20], genre_counts.values[:20], color='#4287f5') 
plt.xlabel('Count')
plt.ylabel('Genre')
plt.title('Top 20 Most Common Genres')
plt.tight_layout()
plt.show()
```

2.What movie got the highest popularity ? what's its genre?

```python
df[df['Popularity'] == df['Popularity'].max()]
```

3.What movie got the lowest popularity ? what's its genre?

```python
df[df['Popularity'] == df['Popularity'].min()]
```
